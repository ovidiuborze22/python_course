```python
# updating and adding new columns
# if you add one value only you will get an error
# because the Length of values (1) does not match length of index (5)
# you can either add 5 values in the list 
# or a simple way is to do a multiplication if the values are the same
# for example North America
df4["Continent"]=df4.shape[0]*["North America"]
df4
```

```python
# updating a column by concatenating two columns and separated by a comma 
df4["Continent"]=df4["Country"]+ "," + "North America"
df4
```

```python
# adding a new row
# transpose data first
df4_t=df4.T
df4_t
```

```python
df4_t["My Address"]=["My City", "My Country",10,7, "My shop", "My State", "My Continent"]
df4_t
```

```python
# to transpose the data back to previous form
# new row will apear at the end
df4=df4_t.T
df4
```
